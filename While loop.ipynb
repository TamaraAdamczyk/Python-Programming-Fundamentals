```python
#Iterating through list dates and stopping at the year 1973, printing out the number of iterations.
dates = [1982,1980,1973,2000]
i = 0
year = dates[0]

while year != 1973:
    print(year)
    i = i + 1
    year = dates[i]

print("It took ", i ,"repetitions to get out of loop.")
```

```python
# Exitting the loop when i is 3
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

```python
# Exitting the loop when i is 3, without printing 3
i = 1
while i < 6:
    if i == 3:
        break
    print(i)
    i += 1
```

```python
# Continue to the next iteration if i is 3
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```
